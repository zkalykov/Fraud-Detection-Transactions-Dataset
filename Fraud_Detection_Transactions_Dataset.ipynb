```python
import os

def show_metrics(df, file_path, name):
    print(f"\n{'='*20} {name} {'='*20}")

    # 1. Rows and Columns
    rows, cols = df.shape

    # 2. Size calculations
    file_mb = os.path.getsize(file_path) / (1024 * 1024)
    ram_mb = df.memory_usage(deep=True).sum() / (1024 * 1024)

    # 3. Output
    print(f"Total Rows:      {rows:,}")
    print(f"Total Columns:   {cols}")
    print(f"Total Data:      {df.size:,} data points")
    print(f"Size on Disk:    {file_mb:.2f} MB")
    print(f"Size in RAM:     {ram_mb:.2f} MB")

# Execute for your two datasets
show_metrics(df_fraud, '/content/synthetic_fraud_dataset.csv', "Fraud Transactions")
show_metrics(df_paysim, '/content/PS_20174392719_1491204439457_log.csv', "PaySim Dataset")
```

```python
import pandas as pd

# 1. Load dataframes (assuming they are already loaded as df_fraud and df_paysim)
# If not, uncomment these lines:
# df_fraud = pd.read_csv('/content/synthetic_fraud_dataset.csv')
# df_paysim = pd.read_csv('/content/PS_20174392719_1491204439457_log.csv')

# 2. Drop unnecessary columns (IDs and redundant flags)
df_fraud_clean = df_fraud.drop(['Transaction_ID', 'User_ID'], axis=1)
df_paysim_clean = df_paysim.drop(['nameOrig', 'nameDest', 'isFlaggedFraud'], axis=1)

# 3. Standardize column names for the merge
# We rename 'Transaction_Amount' and 'amount' to a common 'amount'
# We rename 'Fraud_Label' and 'isFraud' to 'is_fraud'
df_fraud_clean = df_fraud_clean.rename(columns={
    'Transaction_Amount': 'amount',
    'Transaction_Type': 'type',
    'Fraud_Label': 'is_fraud'
})

df_paysim_clean = df_paysim_clean.rename(columns={
    'isFraud': 'is_fraud'
})

# 4. Merge the datasets (Stacking them vertically)
# We use concat because the datasets represent different types of transactions
combined_df = pd.concat([df_fraud_clean, df_paysim_clean], ignore_index=True)

# 5. Review the result
print(f"Merged Dataset Shape: {combined_df.shape}")
print("\n--- Value Counts for the Merged Target Variable (is_fraud) ---")
print(combined_df['is_fraud'].value_counts())

display(combined_df.head())
```

```python
import pandas as pd

# Create a summary table for the merged 'combined_df'
column_summary = []

for col in combined_df.columns:
    column_summary.append({
        "Column Name": col,
        "Data Type": combined_df[col].dtype,
        "Non-Null Count": combined_df[col].count(),
        "Missing Values (%)": f"{(combined_df[col].isna().sum() / len(combined_df) * 100):.2f}%",
        "Unique Values": combined_df[col].nunique(),
        "Sample Value": combined_df[col].dropna().iloc[0] if not combined_df[col].dropna().empty else "All Null"
    })

# Convert to DataFrame for a clean "column-like" display
df_summary = pd.DataFrame(column_summary)

print(f"--- Final Merged Dataset Column Analysis ({combined_df.shape[0]} rows) ---")
display(df_summary)
```

```python
cols_to_keep = ['amount', 'type', 'step', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'is_fraud']
df_train = combined_df[cols_to_keep].copy()
cols_to_fix = ['step', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
df_train[cols_to_fix] = df_train[cols_to_fix].fillna(0)

print(df_train.shape)
print(df_train.isnull().sum())
display(df_train.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 24))

# 1. Distribution of Fraud vs. Non-Fraud (Position 1 of 4)
plt.subplot(4, 1, 1)
sns.countplot(x='is_fraud', data=df_train, palette='viridis', hue='is_fraud', legend=False)
plt.title('Distribution of Fraud vs. Non-Fraud')
plt.yscale('log')

# 2. Transaction Type vs. Fraud (Position 2 of 4)
plt.subplot(4, 1, 2)
sns.countplot(x='type', hue='is_fraud', data=df_train)
plt.title('Transaction Type vs. Fraud')
plt.xticks(rotation=45)
plt.yscale('log')

# 3. Transaction Amount Distribution (Position 3 of 4)
plt.subplot(4, 1, 3)
sns.boxplot(x='is_fraud', y='amount', data=df_train)
plt.title('Transaction Amount by Fraud Label')
plt.yscale('log')

# 4. Correlation Heatmap (Position 4 of 4)
plt.subplot(4, 1, 4)
correlation = df_train.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Data Transformation: One-Hot Encoding the 'type' column
df_final = pd.get_dummies(df_train, columns=['type'], drop_first=True)

# 2. Data Transformation: Normalization (Scaling numeric values)
scaler = StandardScaler()
num_cols = ['amount', 'step', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
df_final[num_cols] = scaler.fit_transform(df_final[num_cols])

print("Transformation Complete!")
display(df_final.head())
```

```python
# 1. Separate the fraud and normal transactions
fraud_data = df_final[df_final['is_fraud'] == 1]
normal_data = df_final[df_final['is_fraud'] == 0]

# 2. Data Reduction (Sampling): Take a random sample of normal data equal to the fraud count
normal_sample = normal_data.sample(n=len(fraud_data), random_state=42)

# 3. Combine them into a balanced training set
df_balanced = pd.concat([fraud_data, normal_sample])

# 4. Shuffle the data so the model doesn't see all fraud first
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Balanced Dataset Shape: {df_balanced.shape}")
print(df_balanced['is_fraud'].value_counts())
```
